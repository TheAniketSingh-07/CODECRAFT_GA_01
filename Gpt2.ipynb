```python
from transformers import pipeline
from IPython.display import display
import ipywidgets as widgets

# Load your fine-tuned model
generator = pipeline("text-generation",
                     model="/content/gpt2-custom",
                     tokenizer="/content/gpt2-custom")

# Create text input box
input_box = widgets.Text(
    value='',
    placeholder='Type your question or prompt here...',
    description='Prompt:',
    layout=widgets.Layout(width='80%')
)

# Create output area
output_area = widgets.Output()

# Define button click function
def on_button_click(b):
    output_area.clear_output()
    prompt = input_box.value.strip()
    if prompt == "":
        with output_area:
            print("‚ùó Please enter a prompt.")
        return
    # Generate output
    response = generator(prompt,
                         max_new_tokens=100,
                         do_sample=True,
                         temperature=0.9,
                         top_k=50,
                         truncation=True)

    with output_area:
        print(f"\nüü¢ Prompt: {prompt}\n")
        print("üìù Response:")
        print(response[0]['generated_text'])

# Create the generate button
generate_button = widgets.Button(
    description='Generate Text',
    button_style='success'
)
generate_button.on_click(on_button_click)

# Display the UI
display(input_box, generate_button, output_area)
```

```python
from transformers import GPT2LMHeadModel, Trainer, TrainingArguments

model = GPT2LMHeadModel.from_pretrained("gpt2")

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=2,
    num_train_epochs=3,
    logging_steps=50,
    save_steps=200,
    warmup_steps=50,
    weight_decay=0.01,
    logging_dir="./logs"
)

# Add labels to the dataset
tokenized = tokenized.add_column("labels", tokenized["input_ids"])

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=tokenized,
    eval_dataset=tokenized,  # You can split for validation
)

trainer.train()
```

```python
from datasets import Dataset
from transformers import AutoTokenizer

# Load dataset from text
with open("Dataset.txt", "r", encoding="utf-8") as f:
    data = f.read().split("=== ")

# Reformat
records = [{"text": "=== " + item.strip()} for item in data if item.strip()]
dataset = Dataset.from_list(records)

# Tokenize
tokenizer = AutoTokenizer.from_pretrained("gpt2")
tokenizer.pad_token = tokenizer.eos_token # Add padding token
def tokenize(batch): return tokenizer(batch["text"], truncation=True, padding="max_length", max_length=512)
tokenized = dataset.map(tokenize, batched=True)
```
